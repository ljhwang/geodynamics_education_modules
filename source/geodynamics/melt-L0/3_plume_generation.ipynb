## Plume generation
In a convecting mantle, we can understand plumes as the transient instabilities that arise from the thermal boundary layer at the base of the mantle. The criteria for onset of localized thermal instability depends on the Rayleigh number, $Ra$, as:

$$Ra = \frac{g \alpha \Delta T (\delta t)^3}{\eta \kappa}$$

where, $g$ is acceleration due to gravity,  $\alpha$ is the thermal expansivity, $\Delta T$ is the superadiabatic temperature difference across the thermal boundary layer, $\delta t$ is the local boundary layer thickness, $\kappa$ is the thermal diffusivity, and $\eta$ is the viscosity of the rising instability.

Assuming that the boundary layer grows through diffusion as $\delta t = \sqrt{\pi \kappa t}$ and that instability occurs when $Ra$ exceeds a critical value, $Ra_c$, we can derive the time to onset of instability as:

$$t_c = \left(\frac{{Ra}_c \eta }{g \alpha \Delta T} \right)^{2/3} \frac{1}{\pi \kappa^{1/3}}$$

Assuming representative values for mantle plume (table 11.3 from Turcotte and Schubert 2001):
- $g = 9.8 \, m/s^2$
- $\alpha = 3 \times 10^{-5} \, K^{-1}$
- $\Delta T = 200 \, K$
- $\kappa = 0.8 \times 10^{-6} \, m^2/s$
- $Ra_c = 3000$ (critical Rayleigh number for onset of instability),

we get $t_c \approx 30 \, Myr$, for a low viscosity of source material. This is short compared to the time scale of mantle convection, which is on the order of hundreds of millions of years, suggesting that conditions are met for localized instabilities to form.

For plumes to develop from these instabilities, the growth rate of instability must be large enough to separate the plume from the boundary layer before it is advected away by mantle flow.

Modeling the plume as a spherical diapir as it ascends through the surrounding mantle, we can compute the terminal ascent speed, $U$, and terminal volume, $V$  of the plume head as:

\begin{align}
U &= \left(\frac{g' Q_v}{8 \pi \nu_p}\right)^{1/2} ;
V &= \frac{4\pi}{3} (3\nu_m)^{3/2} \left(\frac{Q_v}{8 \pi g' \nu_p}\right)^{3/4}
\end{align}

where $g' = g \Delta \rho/\rho$ is the diapir buoyancy, $\Delta \rho$ is the density difference between mantle and the plume, $Q_v = U A$ is the volume flux of fluid through the trailing conduit and A is the cross-sectional area of the plume conduit, and $\nu_p$ and $\nu_m$ is the kinematic viscosity of the plume and mantle, respectively.

### Flood basalt eruptions
Food basalt eruptions are the result of extrusion of melt from the plume head as it is transported into the lithosphere from below.

Using the some flood basalts and hot spot eruption rates:

| Flood basalt | Approximate Age (Myr) | Eruption volume ($10^6 km^3$) | Duration (Myr) | Hot spot |  Average Eruption Rate (km$^3$/yr) |
| --- | --- | --- | --- | --- | --- |
| Deccan Traps | 67 | 1.5 | 1 | Reunion | 0.04 |
| N. Atlantic  | 60 | 2   | 2 | Iceland | 0.02 |
| Parana       | 125| 1.5 | 2 | Tristan | 0.03 |

*Values from Table 11.5 from Turcotte and Schubert (2001)*

The above values suggest that ascent velocity and the volume flux beneath the Renuion hot spot is about $ U = 0.27 m/yr$, and $Qv = 2.9 km^3/yr$. This means that the approximate cross-sectional area is $A = Q_v/U = 1000 km^2$, which corresponds to a plume conduit diameter of about 120 km.
This radius is large enough to explain the horizontal extent of the observed volcanism.

### Impact of volcanism on climate and mass extinctions

As volcanism occurs, it can release large amounts of gases into the atmosphere, including carbon dioxide, sulfur dioxide, and water vapor, known as volatile outgassing. This process can lead to rapid climate change and can lead to mass extinction events. Evidence of such correlation exists between the timing of the known mass extinction events and the continental flood basalt eruptions.
