## Partial Melting and Fractional Crystallization

Due to the heterogeneous mineral composition of the mantle rocks and their different melting points, mantle rocks do not melt uniformly. Instead, partial melting occurs, where only certain minerals melt while others remain solid. This process results in the formation of magmas with distinct compositions.

### Bowen's Reaction Series

Bowen's reaction series describes the temperature at which minerals crystallize or melt when heated. Mg- and Fe-rich minerals (mafic) crystallize at higher temperatures, while Si-, Al- and K-rich minerals (felsic) crystallize at lower temperatures. This sequence help us understand the formation of igneous rocks of different compositions from a common magma source.

### Magma enrichment and depletion

As magma cools, the mafic minerals crystallize first leaving the remaining melt enriched in silica and other elements incompatible with the crystal-structure of the crystallized mafic minerals. This partitioning of magma into melt and residue (i.e., crystallized minerals) is known as fractional crystallization.

### Extrusive and Intrusive Igneous Rocks

If magma cools slowly, allowing large crystals to form, it results in **intrusive** igneous rocks (e.g., gabbro). Conversely, if magma cools rapidly at the surface, it forms **extrusive** igneous rocks (e.g., basalt) with smaller crystals.
