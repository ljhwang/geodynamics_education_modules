## Description<a name="label:desc"></a>

This module introduces the fundamental processes governing subduction dynamics and the role of geodynamic models in constraining these processes.

## Learning objectives

### Scientific
Upon completion of this module you would be familiar with:
- Forces acting on a subducting slab
- Computing frictional heating along the slab interface
- Factors controlling the morphology of the slab as it interacts with the mantle transition zone

### Technical
Upon completion of this module you will be able to:
- Run ASPECT from the command line
- Understand the basic layout of model simulation file
- Visualize model simulation results using Python

## Dependencies
- Python (> version 3.3) with the following libraries installed:
  * Matplotlib
  * NumPy
  * PyVista

- ASPECT (> version 3.0) installed locally or using the ASPECT Jupyter notebook on the CIG HUBZero platform [here](https://geodynamics.org/resources/tools)
