## Estimating the magnitude of the forces

Thermal profile of the slab can be approximated using a half-space cooling model as:

\begin{equation}
T(t, z) = T_m \text{erf} \left( \frac{z}{2\sqrt{\kappa t}} \right)
\end{equation}

where $T_m$ is the mantle temperature, $\kappa$ is the thermal diffusivity, and $t$ is the age of the slab at the trench. For details, see the heat flow module in heat-flow-L0/8_cooling_of_oceanic_lithosphere.ipynb on half-space cooling model.

To simplify the computation, we can solve the problem in 2D by estimating the magnitude of the forces on a very wide slab.

## Forces acting on the subducting slab

Key forces acting on the subducting slab include:
- *Slab pull* ($F_{SP}$): The negative buoyancy of the cold, dense oceanic lithosphere as it sinks into the mantle pulling the attached tectonic plate.

- *Phase changes*: As the slab descends, it undergoes phase changes at a different depth compared to the surrounding mantle, which can either promot or impede sinking.

- *Slab suction* ($F_{SU}$): A net upward suction force on the slab due to the subduction-induced corner flow.

- *Drag forces* ($F_{DF}$): Resistance to the slab's descent at the interface between the slab and the surrounding mantle.

- *Bending resistance* ($F_{BR}$): The resistance to bending of the plate at the trench.

- *Ridge-push* ($F_{RP}$): The force exerted by the elevated mid-ocean ridge pushing the plate away from the ridge.

### Forces due to phase changes

At the exothermic phase transition of olivine to spinel, the slab becomes denser than the surrounding mantle at a shallower depth, which exerts a driving force on the sinking slab. We can estimate this force by considering the density of the material between the elevated and normal phase transition depth as:

\begin{align*}
F_{Ph} = g \int_{z' = 0}^\infty  \Delta \rho_{ph} h(z') dz' = g \Delta \rho_{ph} \int_{z' = 0}^\infty \frac{\gamma (T_m - T (z'))}{\rho_0 g} dz'  
\end{align*}

where $\gamma \equiv \frac{dP}{dT}$ is the Clapeyron slope of the phase transition.

Substituting the slab temperature (1) above and integrating, we have:

\begin{align*}
F_{ph} =  \frac{2 \gamma \Delta \rho_{ph} T_m}{\rho_0} \sqrt{\kappa t/\pi}. \tag 3
\end{align*}

### Slab pull

The negative buoyancy of  the slab inside the mantle using Archimedes' principle is:

\begin{align*}
F_{SP} &= g \int_A  \Delta dA = g \rho_0 \alpha \int_A  (T_m - T) dA ,
\end{align*}

where $T$ is the slab temperature, $\alpha$ is the thermal expansivity, and $\rho_0$ is the reference mantle density.

Let's assume the total length of slab $b$ inside the mantle, and $z'$ is the direction perpendicular to the slab surface. Substituting the slab temperatures and integrating the total force from $z'=0$ to $z' \rightarrow \infty $:

\begin{align*}
F_{SP} &= g \rho_0 \alpha b \int_{z' = 0}^\infty  T_m - T_m \text{erf} \left( \frac{z'}{2\sqrt{\kappa t}} \right)  dz'  = 2 g \rho_0 \alpha b T_m \sqrt{\kappa t/\pi}, \tag 2
\end{align*}

using $\int_0^\infty \text{erfc}(x) dx = 1/\sqrt{\pi}$. Outside the slab, the temperatures are $T = T_m$, and does not contribute anything to the force integral.

> **Note**: We can similarly compute the phase change force for an endothermic reaction (e.g., spinel to perovskite), which has a negative Clapeyron slope, impeding the slab descent into the lower mantle.

### Ridge push
The net horizontal force acting on the slab due to gravitational body forces at the ridge and away from it can be illustrated below:  

<div>
<img src='./images/ridge_push_force.png' width="600"/>
<figcaption align = "center"> Schematic illustration of horizontal forces acting at ocean ridge. Image: Turcotte and Schubert (2002) </figcaption>
</div>

In the above figure, $F_4 = F_2$  and $F_3$ are the pressure forces exerted by the column of water and lithosphere, respectively, while $F_5 = F_1$ is the total pressure force acting on the lithosphere from the surrounding mantle.  

The ridge push force is then:

\begin{align*}
F_{RP} = F_5 - F_4 - F_3 = \int_0^{ w + yL} \rho_m g y dy - \int_0^w \rho_w g y dy - \int_{z=0}^{w + L} (\rho_w g W +  \rho_L g dz)dz,
\end{align*}

where the density of the plate, $\rho_L$, which depends on the thermal structure of the slab (1). Integrating the above equation:

\begin{align*}
F_{RP} = g \rho_m \alpha T_m \left[1 + \frac{2 \rho_m \alpha T_m}{\pi (T_m - T_w)} \right] \kappa t \tag 4
\end{align*}

See Turcotte and Schubert (2002) for details.

### Drag Force

Viscous shear forces acting on a slab as it subducts into mantle can be modeled assuming a linear velocity gradient on the sides of the slab (i.e., strain rate) as $2 v_{slab}/w_{slab}$ assuming the slab velocity and width as $v_{slab}$ and $w_{slab}$, respectively:

\begin{align*}
F_{DF} = \frac{2 v_{slab} (\eta_m z_m)}{w_{slab}} \tag 5
\end{align*}

Here, $\eta_m$ is the mantle viscosity and $z_m$ is the length of slab inside mantle.

### Force Magnitudes

Using typical values for the slab–plate system in equation (2), (3), and (4):

- $T_m = 1200 K$
- $\rho_0 = 3300 kg/m^3$
- $\alpha = 3 \times 10^{-5} K^{-1}$
- $\kappa = 10^{-6} m^2/s$
- $t = 80 Ma$
- $\gamma = 4 MPa/K$
- $\Delta \rho_{ph} = 300 kg/m^3$

 $F_{SB} = 4.04 \times 10^{13}$ N/m for a 600 km long slab,
 $F_{ph} = 2.4 \times 10^{13} $ N/m,
and $F_{RP} = 3.1 \times 10^{12}$ N/m.

These values suggest that ridge push is approximately an order of magnitude smaller than the slab pull force. In other words, slabs are primary drivers of plate tectonics.


Using equation (5) and typical values of an old slab subducted in upper mantle, we can compute approximate drag forces on the slab:

- $v_{slab} = 2 $ cm/yr
- $w_{slab} = 80 $ km
- $\eta_m   = 1e21 $ Pa.s
- $z_m      = 600 $ km

$F_{DF}\sim1 \times 10^{13}$ N/m .

The above resistive force is similar in magnitude to the dirving forces computed above. We did not account for bending resistance here and other forces that may impede slab sinking, e.g., endothermic phase change at 660 km.

Under no slab acceleration, the net driving forces acting on the slab are balanced by the resisting forces.
