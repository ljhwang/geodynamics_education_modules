## Slab Morphology
The endothermic phase change at 660 km may impede the slab sinking through from the upper mantle into the lower mantle, affecting the slab's structure as it interacts with the mantle transition zone (MTZ).

The most detailed information we have of slab structure comes from seismic tomography (see figure below). Several factors influence the slab morphology including its strength, age, trench motion (advance or rollback), viscosity jump at the MTZ, and Clapeyron slope of the phase transition.

### Trench motion
Slab morphology at the MTZ also affects the surface trench rollback/advance motions. Integrated over the last 50 Ma, most flattened slabs have undergone trench retreat (see figure below).

<img src='./images/trench_rollback.png' width="500"/>

</div>

Subduction dynamics that govern the trench motion can be understood using numerical models.

### Dynamic models of subducting slab

#### Influence of slab age
Dynamic models show that older slabs are denser and more resistant to bending, which lead to stronger trench retreat than the younger slabs.  

Moreover, models by Garel et al. (2014) suggest that older overriding plates inhibit trench retreat as they offer more resistance to stretching induced from the slab pull and associated corner flow generated in the mantle wedge.

#### Influence of 660 km phase change

Agrusta et al. (2017) characterize slab's dynamics for varying Clapeyron slope and viscosity jump values at the 660 km boundary (see figure below).

At high Clapeyron slopes, slabs of all ages are likely to stagnate while at low Clapeyron slope values, slabs of all ages are likely to penetrate if the viscosity jump to lower mantle is not too high.
Increasing the viscosity of lower mantle, causes some older slabs to stagnate even at low Clapeyron values.
